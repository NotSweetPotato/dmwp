```python
def create_cir():
    circle = plt.Circle((0, 0), radius=0.05)
    return circle

def show_shp(patch):
    ax = plt.gca()
    ax.add_patch(patch)
    plt.axis('scaled')
    plt.show()

show_shp(create_cir())
```

```python
from IPython.display import HTML
from matplotlib import animation

def update_radius(i, circle):
    circle.radius = i * 0.5
    return circle

def create_anim():
    fig = plt.gcf()
    ax = plt.axes(xlim=(-10,10), ylim=(-10,10))
    ax.set_aspect('equal')
    circle = create_cir()
    ax.add_patch(circle)
    anim = animation.FuncAnimation(fig, update_radius, fargs = (circle,), frames=30, interval=50)
    plt.close(fig)
    return anim

anim = create_anim()
HTML(anim.to_jshtml())
```

```python
import math

g = 9.8

def get_intervals(u, theta):
    t_flight = 2 * u * math.sin(theta) / g
    intervals = []
    start = 0
    interval = 0.005
    while start < t_flight:
        intervals.append(start)
        start += interval
    return intervals

def update_pos(i, circle, intervals, u, theta):
    t = intervals[i]
    x = u * math.cos(theta) * t
    y = u * math.sin(theta) * t - 0.5 * g * t * t
    circle.center = x, y
    return circle

def ball_animation(u, theta):
    theta = math.radians(theta)
    intervals = get_intervals(u, theta)

    fig = plt.gcf()
    ax = plt.axes(xlim=(0, 20), ylim=(0, 20))
    ax.set_aspect('equal')
    circle = plt.Circle((0, 0), 1)
    ax.add_patch(circle)

    anim = animation.FuncAnimation(fig, update_pos, frames=len(intervals),
                                   fargs=(circle, intervals, u, theta), interval=1)
    plt.close(fig)
    return anim

anim = ball_animation(15, 70)
HTML(anim.to_jshtml())
    
```

```python
import random

def trans_1(p):
    return p[0] + 1, p[1] - 1

def trans_2(p):
    return p[0] + 1, p[1] + 1

def trans(p):
    return random.choice([trans_1, trans_2])(p)

def draw_1(num):
    p = [1, 1]

    p_list = [p]
    for i in range(num):
        p_list.append(trans(p_list[-1]))
    
    x = [pp[0] for pp in p_list]
    y = [pp[1] for pp in p_list]
    plt.plot(x, y)
    plt.show()

draw_1(100)
```

```python
def trans_1(p):
    x = 0.85 * p[0] + 0.04 * p[1]
    y = -0.04 * p[0] + 0.85 * p[1] + 1.6
    return x, y

def trans_2(p):
    x = 0.2 * p[0] - 0.26 * p[1]
    y = 0.23 * p[0] + 0.22 * p[1] + 1.6
    return x, y

def trans_3(p):
    x = -0.15 * p[0] + 0.28 * p[1]
    y = 0.26 * p[0] + 0.24 * p[1] + 0.44
    return x, y

def trans_4(p):
    x = 0
    y = 0.16 * p[1]
    return x, y

def trans(p):
    probs = [0.85, 0.07, 0.07, 0.01]
    r = random.random()
    i = 0
    s = probs[0]
    while s < r and i < len(probs):
        i += 1
        s += probs[i]

    trans_list = [trans_1, trans_2, trans_3, trans_4]
    return trans_list[i](p)

def draw(num):
    p = [0, 0]

    p_list = [p]
    for i in range(num):
        p_list.append(trans(p_list[-1]))
    
    x = [pp[0] for pp in p_list]
    y = [pp[1] for pp in p_list]
    plt.scatter(x, y)
    plt.show()

draw(2000)
```

```python
import matplotlib.cm as cm

def init_img(x_p, y_p):
    img = []
    for i in range(y_p):
        x_c = []
        for j in range(x_p):
            x_c.append(0)
        img.append(x_c)
    return img

def color_pnts():
    x_p, y_p = 6, 6
    img = init_img(x_p, y_p)
    for i in range(y_p):
        for j in range(x_p):
            img[i][j] = random.randint(0, 10)
    plt.imshow(img, origin='lower', extent=(0, 5, 0, 5), cmap=cm.Greys_r, interpolation='nearest')
    plt.colorbar()
    plt.show()

color_pnts()
```
